# hypothesis testing
1. Tagihan kesehatan perokok lebih tinggi daripada tagihan kesehatan non perokok
2. Tagihan kesehatan dengan BMI diatas 25 lebih tinggi daripada tagihan kesehatan
dengan BMI dibawah 25

```python
from scipy.stats import t
```

## 1
**tagihan kesehatan perokok lebih tinggi daripada tagihan kesehatan non perokok**

1. null hypothesisnya tagihan perokok > tagihan non perokok
alternative hypothesisnya tagihan perokok !> tagihan non perokok
significance level 0.05
2. pake t test karena basically bandingin 2 sampel: perokok sama non perokok
3. aturan keputusannya pake two-tailed
kalo p-value < alpha, H0 kita reject
kalo nilai kritis lower tail > statistik uji > nilai kritis di upper tail, H0 kita reject
4. hitung statistik ujinya pake yg variansnya beda

#### 1. hypothesis and significance level

- $H_0$: tagihan perokok $\leq$ tagihan non-perokok
- $H_1$: tagihan perokok > tagihan non-perokok

```python
alpha = 0.05
```

#### 2. statistik uji

```python
len(insurance)
```

```python
len(smoker)
```

```python
len(nsmoker)
```

- statistik uji: t-test, karena membandingkan 2 group

#### 3. aturan keputusan
- lower tailed
- nilai uji < critical value lower-tailed, maka H0 ditolak

#### 4. hitung statistik uji

```python
print(f'{np.var(smoker.charges.sample(n=200)):.0f}, {np.var(nsmoker.charges.sample(n=200)):.0f}')
```

```python
from scipy.stats import ttest_ind
```

```python
stat, p = ttest_ind(a = smoker.charges.sample(n=50), b = nsmoker.charges.sample(n=50), equal_var=False, alternative='greater') # eaual_var= False karena varians kedua populasi berbeda

# Interpretasi Hasil
print('Statistics = %.4f, p-value = %.4f' % (stat, p)) 
```
